```python
new_df = df.copy()

# Replace 5% and 10% with 13%
new_df['discount'] = new_df['discount'].replace({'5%': '13%', '10%': '13%'})
print(new_df)
```

```python
new_df = df.copy()

# mapping for categorical to numerical values
rating_map = {
    'Excellent': 4,
    'very Good': 3,
    'Good': 2,
    'Average': 1
}
new_df['rating'] = new_df['rating'].replace(rating_map)

print(new_df)
```
